# 1) Technical Analysis of SP500

This will include a backtest and suggest some refinements.

The analysis is to use 2 rolling averages - 2 months and 1 year and is based on the observstion that the yearly rolling average reacts much slower to daily prices changes than a 2 month rolling average.
Using this knowledge, devise a simpel trading rule / regime that will sign buy/sell/hold signals.
Compare the trading regime against market performance

## 1.1) Load in the Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 1.2) Load in the data

You will load in the index information from  csv file as you dont have access to live data.

However, this type of information is easy to get from an online source - e.g. Bloomberg API.

Also, notice here that you need to convert string types to numeric types - this is quite often the case when reading date direct from file.

```python
df_SP500 = pd.read_csv('SP500Historical.csv', index_col='Date', parse_dates=True)

df_SP500 = df_SP500.sort_index(ascending=True)

# Need to change strings to numerics - quite often needed to be done when loading data from files
df_SP500['Price'] =pd.to_numeric(df_SP500['Price'].str.replace(',',''))
df_SP500['Open'] =pd.to_numeric(df_SP500['Open'].str.replace(',',''))
df_SP500['High'] =pd.to_numeric(df_SP500['High'].str.replace(',',''))
df_SP500['Low'] =pd.to_numeric(df_SP500['Low'].str.replace(',',''))

# Just checking the dates are the right way round - important to do this as they might need sorted and / or re-indexed
df_SP500.index
```

## 1.3) Calculate the rolling averages

2 average in this piece of analysis here
- a 42 day rolling average (approx 2 months) - follows the daily movements very closely.
- a 252 day rolling average (approx 1 year) - much slower to react to daily movements.

```python
# Based on a 2 month (42 day) and a 1 year (252 trading day)
x42 = df_SP500['Price'].rolling(42).mean()
x252 = df_SP500['Price'].rolling(252).mean()

df_SP500['42d'] = np.round(x42,2)
df_SP500['252d'] = np.round(x252,2)
```

## 1.4) A simple plot to aid understanding of whats going on

```python
df_SP500[['Price', '42d', '252d']].plot(figsize=(18,9))
```

## 1.5) Create more data points

Here we take the difference between the 2 months rolling average and the yearly rolling average.

Add a new column to the DataFrame called 42d-252d and populate it with the diference between out 2 rollign averages

```python
df_SP500['42d-252d'] = df_SP500['42d'] - df_SP500['252d']
df_SP500.tail()
```

## 1.6) Plot the Regime

Again a quick eyeball of what we have done aids understanding.

This plot is the beginning of how we generate our buy/sell/hold signals


```python
df_SP500['42d-252d'].plot(grid=True)
```

## 1.7) Add the rules of a Trading Regime

Add a new column called 'Regime'

The trading regime here is based on the following rule.

- Generate a LONG signal when the 42-252 day difference first passes the **3% threshhold ** 
- Generate a SHORT signal when the 42-252 day difference first passes the ** - 3% threshhold ** 
- Generate a HOLD signal (convert to cash) when the 42-252 day difference first enters the ** +/- 3% band ** 

The encoding for the Regime is
- 1 ==> Long
- -1 ==> Short
- 0 ===> Hold

Note the use of **np.where** function here - there are many other ways to do this but the vector arithmetic in numpy fits extremely well with the rows and columns of pandas DataFrames.


```python
# 3 %
threshold = 0.03
    
df_SP500['Regime'] = np.where(df_SP500['42d-252d'] / df_SP500['252d'] > threshold, 1, 0)
df_SP500['Regime'] = np.where(df_SP500['42d-252d'] / df_SP500['252d'] < -threshold, -1, df_SP500['Regime'])
```

## 1.8) Plot the trading signals

This plot basically tells us when to go Long, Short or Hold.

e.g. It says Hod in cash from start of 2006.<br>
Go LONG around early 2007 and stay there im ner enf of 2007<br>
Go to HOLD for a short period, then go LONG, then go to HOLD shorly after.<br>
etc etc etc

over the oeriod 2006 to 2018 - this strategy has generated about 20 trading signals

```python
df_SP500['Regime'].plot(figsize=(18,6))
```

## 1.9) Back test

2 new columns 
- 'Market' - how the SP500 performed during this period
- 'Strategy' = how our trading strategy performed during this period

Use log of differences to calculate performance of the market
Our strategy's performance is the prodct of market performance and our regime.

```python
df_SP500['Market'] = np.log(df_SP500['Price'] / df_SP500['Price'].shift(1))
df_SP500['Strategy'] = df_SP500['Regime'].shift(1) * df_SP500['Market']
```

## 1.10) Plot the results

A simple plot of market vs strategy shows how well the strategy perfomred against the market.

Calculate a continuous accumulation of the returns (np.exp)

The strategy 
- slightly underperforms at the start of the analysis period
- significantly out performs the market between middle 2008 thru to start of 2013
- slightly outperforms the market significantly during form 2013 to 2016
- slighly underperfoms between 2016 to present day


```python
df_SP500[['Market', 'Strategy']].cumsum().apply(np.exp).plot(grid=True)
```
